```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('horse.jpg') #BGR
show_horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) #RGB

rainbow = cv2.imread('rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_bricks)
```

```python
#OPENCV BGR
hist_values = cv2.calcHist([blue_bricks],
             channels=[0],
             mask=None,
             histSize=[256],
             ranges=[0,256])
```

```python
saruman = cv2.imread('saruman.jpg',0)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(saruman)
```

```python
saruman.shape
```

```python
hist_values = cv2.calcHist([saruman],
                          channels=[0],
                          mask=None,
                          histSize=[256],
                          ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_saruman = cv2.equalizeHist(saruman)
```

```python
display_img(eq_saruman)
```

```python
hist_values = cv2.calcHist([eq_saruman],
                          channels=[0],
                          mask=None,
                          histSize=[256],
                          ranges=[0,256])
```

```python
plt.plot(hist_values)
```
