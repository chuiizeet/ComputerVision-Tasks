```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chessboard = cv2.imread('real_chessboard.jpg')
real_chessboard = cv2.cvtColor(real_chessboard,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chessboard)
```

```python
gray_real_chess = cv2.cvtColor(real_chessboard,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray,
                      blockSize=2,
                      ksize=3,
                      k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()] = [0,255,0] #RGB
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)

dst = cv2.cornerHarris(src=gray,
                      blockSize=2,
                      ksize=3,
                      k=0.04)

```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chessboard[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(real_chessboard)
```

```python
#Reload images

flat_chess = cv2.imread('flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

real_chessboard = cv2.imread('real_chessboard.jpg')
real_chessboard = cv2.cvtColor(real_chessboard,cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_real_chess = cv2.cvtColor(real_chessboard,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,
                                  64,
                                  0.01,
                                  10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,
               (x,y),
               3,
               (255,0,0),
               -1)
    
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,
                                  100,
                                  0.01,
                                  10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chessboard,
               (x,y),
               3,
               (255,0,0),
               -1)
    
```

```python
plt.imshow(real_chessboard)
```
