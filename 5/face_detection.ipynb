```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
frodo = cv2.imread('frodo.jpg',0)
dennis = cv2.imread('Denis_Mukwege.jpg',0)
solvay = cv2.imread('solvay_conference.jpg',0)
```

```python
plt.imshow(solvay, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,
                      (x,y),
                      (x+w,y+h),
                      (255,0,0),
                      10)
    return face_img
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
def adj_etect_face(img):
    
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img,
                                              scaleFactor=1.2,
                                              minNeighbors=5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,
                      (x,y),
                      (x+w,y+h),
                      (255,0,0),
                      10)
    return face_img
```

```python
result = adj_etect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
    eyes_rects = eye_cascade.detectMultiScale(face_img,
                                              scaleFactor=1.2,
                                              minNeighbors=5)
    
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(face_img,
                      (x,y),
                      (x+w,y+h),
                      (255,0,0),
                      10)
    return face_img
```

```python
result = detect_eyes(frodo)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result = detect_eyes(dennis)
plt.imshow(result,cmap='gray')
```

```python
cap = cv2.VideoCapture(0)

while True:
    
    ret,frame = cap.read(0)
    
    frame = detect_face(frame)
    cv2.imshow('Video face detection', frame)
    
    k = cv2.waitKey(1)
    if k == 27:
        break
        
cap.release()
cap.destroAllWindows()
```
