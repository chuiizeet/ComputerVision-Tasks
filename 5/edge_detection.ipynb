```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,
                 threshold1=127,
                 threshold2=127)
```

```python
plt.imshow(edges)
```

```python
mid_val = np.median(img)
```

```python
mid_val
```

```python
# Lower threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0,0.7*mid_val))
# Upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255,1.3*mid_val))
```

```python
edges = cv2.Canny(image=img,
                 threshold1=lower,
                 threshold2=upper+150)
plt.imshow(edges)
```

```python
upper
```

```python
blurred_img = cv2.blur(img,
                      ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img,
                 threshold1=lower,
                 threshold2=upper+100)
plt.imshow(edges)
```
