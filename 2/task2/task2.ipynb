```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import copy
%matplotlib inline
```

```python
img = cv2.imread('dog_backpack.jpg')
```

```python
plt.imshow(img)
np.shape(img)
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
t_img = fix_img.copy()
plt.imshow(fix_img)
```

```python
flip_img = cv2.flip(fix_img,0)
plt.imshow(flip_img)
```

## Draw a empty RED rectangle

```python
cv2.rectangle(fix_img,
              pt1=(200,375),
              pt2=(650,700),
              color=(255,0,0),
              thickness=10)
plt.imshow(fix_img)
```

## Draw a BLUE triangle

```python
#t_img
vertices = np.array([[250,650],[650,650],[450,400]], dtype=np.int32)
pts = vertices.reshape((-1,1,2))
cv2.polylines(t_img,[pts], isClosed=True, color=(0,0,255),thickness=20)
```

```python
plt.imshow(t_img)
```

## Bonus: Fill the triangle.

```python
#f_img
f_img = t_img.copy()
cv2.fillPoly(f_img,[pts],color=(0,0,255))
plt.imshow(f_img)
```
